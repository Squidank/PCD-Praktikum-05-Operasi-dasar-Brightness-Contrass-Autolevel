```python
# mengimport library yg dibutuhkan
import numpy as np #  operasi numerik dan manipulasi array.
import imageio #  membaca dan menulis gambar dalam berbagai format.
import matplotlib.pyplot as plt # membuat plot dan visualisasi data
```

## Membaca gambar

```python
img = imageio.imread("image/03.jpg") # membaca gambar di file direktori image menggunakan fungsi imageio
```

## Mendapatkan resolusi dan type dari gambar

```python
# untuk memperoleh informasi gambar
img_height = img.shape[0] # ketinggian
img_width = img.shape[1] # lebar
img_channel = img.shape[2] # channel
img_type = img.dtype # type gambar
```


## Brightness Grayscale

```python
img_brightness = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong dengan ukuran yg sama untuk gambar baru
```

```python
def brighter(nilai):
    for y in range(0, img_height): # Melakukan iterasi untuk setiap baris (y) dalam gambar
        for x in range(0, img_width): # Melakukan iterasi untuk setiap kolom (x) dalam gambar
            
            # Mengambil nilai warna merah (red), hijau (green), dan biru (blue) dari gambar pada posisi (x,y)
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            
            gray = (int(red) + int(green) + int(blue)) / 3 # # Menghitung nilai keabuan (gray) dengan menjumlahkan nilai warna merah, hijau, dan biru, lalu membaginya dengan 3
            gray += nilai # Menambahkan nilai keabuan (gray) dengan nilai yang diberikan (nilai)
            
            # Memastikan nilai keabuan (gray) tidak melebihi 255 (batas maksimum warna)
            if gray > 255:
                gray = 255
            
            # Memastikan nilai keabuan (gray) tidak kurang dari 0 (batas minimum warna)
            if gray < 0:
                gray = 0
                
            img_brightness[y][x] = (gray, gray, gray)# Mengatur piksel gambar pada posisi (y, x) dengan nilai keabuan (gray) yang telah diubah
```

```python
brighter(-100) # menguarangi tingkat brightnes sebesar 100 piksel
plt.imshow(img_brightness) # menampilkan gambar img_brightness
plt.title("Brightness -100") # menambahkan judul
plt.show() # menampilkan bingkai dengan plot subplot yang sudah disiapkan

brighter(100) # menambahkan tingkat brightnes sebesar 100 piksel
plt.imshow(img_brightness)  # menampilkan gambar img_brightness
plt.title("Brightness 100") # menambahkan judul
plt.show() # menampilkan bingkai dengan plot subplot yang sudah disiapkan
```

## Brightness RGB

```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong dengan ukuran yg sama untuk gambar baru
```

```python
def rgbbrighter(nilai):
    for y in range(0, img_height): # Melakukan iterasi untuk setiap baris (y) dalam gambar
        for x in range(0, img_width): # Melakukan iterasi untuk setiap kolom (x) dalam gambar
            red = img[y][x][0]# Mengambil nilai warna merah (red) dari gambar pada posisi (y, x)
            red += nilai  # Menambahkan nilai kecerahan (nilai) ke nilai warna merah (red)
            
            # Memastikan nilai warna merah (red) tidak melebihi 255 (batas maksimum warna)
            if red > 255:
                red = 255
                
            # Memastikan nilai warna merah (red) tidak kurang dari 0 (batas minimum warna)
            if red < 0:
                red = 0
                
            green = img[y][x][1] # Mengambil nilai warna hijau(green) dari gambar pada posisi (y, x)
            green += nilai # Menambahkan nilai kecerahan (nilai) ke nilai warna hijau(green)
            
            # Memastikan nilai warna hijau(green) tidak melebihi 255 (batas maksimum warna)
            if green > 255:
                green = 255
                
            # Memastikan nilai warna hijau(green) tidak kurang dari 0 (batas minimum warna)
            if green < 0:
                green = 0
            
            blue = img[y][x][2] # Mengambil nilai warna biru (blue) dari gambar pada posisi (y, x)
            blue += nilai # Menambahkan nilai kecerahan (nilai) ke nilai warna biru (blue)
            
            # Memastikan nilai warna biru (blue) tidak melebihi 255 (batas maksimum warna)
            if blue > 255:
                blue = 255
                
            # Memastikan nilai warna biru (blue) tidak kurang dari 0 (batas minimum warna)   
            if blue < 0:
                blue = 0
            
            # Mengatur piksel gambar pada posisi (y, x) dengan nilai warna yang telah diubah
            img_rgbbrightness[y][x] = (red, green, blue)
```

```python
rgbbrighter(-100) # menguarangi tingkat brightnes sebesar 100 piksel
plt.imshow(img_rgbbrightness)# menampilkan gambar img_brightness
plt.title("Brightness -100") # menambahkan judul
plt.show() # menampilkan bingkai dengan plot subplot yang sudah disiapkan

rgbbrighter(100) # menambahkan tingkat brightnes sebesar 100 piksel
plt.imshow(img_rgbbrightness)# menampilkan gambar img_brightness
plt.title("Brightness 100")# menambahkan judul
plt.show()# menampilkan bingkai dengan plot subplot yang sudah disiapkan
```

## Contrass Autolevel

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong dengan ukuran yg sama untuk gambar baru
```

```python
def autocontrass():
    
    # Inisialisasi nilai maksimum dan minimum dan nilai d untuk digunakan dalam perhitungan
    xmax = 300 
    xmin = 0
    d = 0 # tingkat contras

    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            
            # Mengambil nilai warna merah (red), hijau (green), dan biru (blue) dari gambar pada posisi (y, x)
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            
            # Menghitung nilai keabuan (gray) dengan menjumlahkan nilai warna merah, hijau, dan biru, lalu membaginya dengan 3
            gray = (int(red) + int(green) + int(blue)) / 3
            
             # Memperbarui nilai maksimum dan minimum jika ditemukan nilai keabuan yang lebih kecil atau lebih besar
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
                
    d = xmin-xmax # Menghitung nilai d sebagai selisih antara nilai minimum dan maksimum
    
    # Mengubah tingkat kontras gambar berdasarkan nilai d dan rumus yang diberikan
    for y in range(0, img_height):
        for x in range(0, img_width):
            # Mengambil nilai warna merah (red), hijau (green), dan biru (blue) dari gambar pada posisi (y, x)
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
        
            # Menghitung nilai keabuan (gray) dengan menjumlahkan nilai warna merah, hijau, dan biru, lalu membaginya dengan 3
            gray = (int(red) + int(green) + int(blue)) / 3
        
            # Menghitung nilai keabuan (gray) yang telah diubah menggunakan rumus yang diberikan
            gray = int(float(255 / d) * (gray - xmax))
        
            # Mengatur piksel gambar pada posisi (y, x) dengan nilai keabuan (gray) yang telah diubah
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
autocontrass() # Mengaplikasikan kontras otomatis pada gambar
plt.imshow(img_autocontrass) # Menampilkan gambar dengan menggunakan matplotlib
plt.title("Contrass Autolevel") # menambah judul
plt.show() # menampilkan bingkai dengan plot subplot yang sudah disiapkan
```
