```python
# mengimport library-library yang dibutuhkan
import matplotlib.pyplot as plt # membuat plot dan visualisasi data
%matplotlib inline
from matplotlib import image # membaca dan menulis gambar
from skimage.io import imread # membaca gambar dari file
```

```python
img = image.imread('image/01.jpg') # membaca file 01 di directory image
img2 = image.imread('image/02.jpg') # membaca file 02 di directory image

# imgCropped = img[0:256,64:320] # crop gmbar salinan dari rentang sumbu x = 64:320, y = 0:256
# img2Cropped = img2[64:256,128:320] # crop gmbar salinan dari rentang sumbu x = 64:256, y = 128:320
imgCropped = img.copy() # menyalin gambar img dari gambar asli
img2Cropped = img2.copy() # menyalin gambar img2 dari gambar asli
imgCropped = imgCropped[0:256,64:320] # crop gmbar salinan dari rentang sumbu x = 64:320, y = 0:256
img2Cropped = img2Cropped[206:560,50:390] # crop gmbar salinan dari rentang sumbu x = 50:390, y = 206:560

print('img Crop Shape : ',imgCropped.shape) # menampilkan ukuran gmbr salinan img setelah di crop
print('img2 Crop Shape : ',img2Cropped.shape) # menampilkan ukuran gmbr salinan img2 setelah di crop


fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # fungsi matplotlib, subplot untuk membuat bingkai (2kolom dan 2baris) figsize untuk mengatur ukuran bingkai)
ax = axes.ravel() # meratakan array menjadi 1 dimensi

ax[0].imshow(img) # menampilkan gambar di bingkai 1
ax[0].set_title("Citra Input 1") # menambah judul

ax[1].imshow(img2) # menampilkan gambar di bingkai 2
ax[1].set_title('Citra Input 2') # menambah judul

ax[2].imshow(imgCropped) # menampilkan gambar di bingkai 3 yang telah di crop sebelumnya
ax[2].set_title("Citra Output 1") # menambah judul

ax[3].imshow(img2Cropped, cmap='gray') # menampilkan gambar di bingkai 4 yang telah di crop sebelumnya dgn memakai colormap gray
ax[3].set_title('Citra Output 2') # menambah judul
```

```python
from skimage.color import rgb2gray #  mengonversi gambar berwarna (RGB) menjadi gambar grayscale.
from skimage.util import invert # membalikkan (invert) piksel dalam gambar. 

import numpy as np # operasi numerik dan manipulasi array.
```

```python
inv = invert(img) # menginvert piksel/warna gambar menjadi gambar negativ
print('Shape Input : ', img.shape) # melihat ukuran gambar
print('Shape Output : ', inv.shape) # melihat ukuran gambar yg telah di invert

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # fungsi matplotlib, subplot untuk membuat bingkai (2kolom dan 2baris) figsize untuk mengatur ukuran bingkai)
ax = axes.ravel()# meratakan array menjadi 1 dimensi

ax[0].imshow(img) # menampilkan gambar di bingkai 1
ax[0].set_title("Citra Input") # menambah judul

ax[1].hist(img.ravel(), bins=256) # menampilkan gambar di bingkai 2, fungsi ravel untuk meratakan array sehingga didapat array pixel tunggal untuk menghitung histogram dengan bins sebagai rentang piksel 8bit
ax[1].set_title('Histogram Input') # menambah judul

ax[2].imshow(inv) # menampilkan gambar di bingkai 3, fungsi ravel untuk meratakan array sehingga didapat array pixel tunggal untuk menghitung histogram dengan bins sebagai rentang piksel 8bit
ax[2].set_title('Citra Output (Inverted Image)') # menambah judul

ax[3].hist(inv.ravel(), bins=256) # menampilkan gambar di bingkai 4, fungsi ravel untuk meratakan array sehingga didapat array pixel tunggal untuk menghitung histogram dengan bins sebagai rentang piksel 8bit
ax[3].set_title('Histogram Output') # menambah judul
```

```python
imggray = rgb2gray(img2Cropped) # mengkonversi gambar rgb menjadi gray

print('img2gray shape = ',imggray.shape) # melihat ukuran gambar

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) # fungsi matplotlib, subplot untuk membuat bingkai (2kolom dan 1baris) figsize untuk mengatur ukuran bingkai)
ax = axes.ravel() # meratakan array menjadi 1 dimensi

ax[0].imshow(img2Cropped) # menampilkan gambar di bingkai 1
ax[0].set_title("img2 RGB Image") # menambah judul
ax[1].imshow(imggray,cmap=plt.cm.gray) # menampilkan gambar gray di bingkai 2, dengan color map gray
ax[1].set_title("img2 Grayscale Image") # menambah judul

fig.tight_layout() # mengatur tata letak yang rapi antara subplot-subplot dan menampilkan bingkai 
plt.show() # menampilkan bingkai dengan plot subplot yang sudah disiapkan
```

```python
copyCamera = imggray.copy().astype(float) # membuat salinan gambar gray dan konversi type data menjadi float

m1,n1 = copyCamera.shape # mendapatkan ukuran/dimensi (m1, n1) dari gambar
output1 = np.empty([m1, n1]) # membuat array kosong output1 dengan dimensi yang sama

for baris in range(0, m1-1): # loop bersarang, melakukan iterasi setiap piksel
    for kolom in range(0, n1-1):
        a1 = baris # Setiap piksel diakses dengan menggunakan indeks (baris, kolom)
        b1 = kolom 
        output1[a1, b1] = copyCamera[baris, kolom] + 10 # meningkatkan intensitas sebesar 10 piksel pada gambar. rumus brightnes fo(x,y)=fi(x,y) + brightnes
        
fig, axes = plt.subplots(2, 2, figsize=(12, 12))# fungsi matplotlib, subplot untuk membuat bingkai (2kolom dan 2baris) figsize untuk mengatur ukuran bingkai)
ax = axes.ravel() # meratakan array menjadi 1 dimensi

ax[0].imshow(copyCamera, cmap='gray') # menampilkan gambar di bingkai 1 dengan colormap gray
ax[0].set_title("Citra Input") # menambah judul

ax[1].hist(copyCamera.ravel(), bins=256) # menampilkan gambar di bingkai 2 fungsi ravel untuk meratakan array sehingga didapat array pixel tunggal untuk menghitung histogram dengan bins sebagai rentang piksel 8bit
ax[1].set_title('Histogram Input') # menambah judul

ax[2].imshow(output1, cmap='gray') # menampilkan gambar di bingkai 3 dengan colormap gray
ax[2].set_title('Citra Output (Brightnes)') # menambah judul

ax[3].hist(output1.ravel(), bins=256) # menampilkan gambar di bingkai 4, fungsi ravel untuk meratakan array sehingga didapat array pixel tunggal untuk menghitung histogram dengan bins sebagai rentang piksel 8bit
ax[3].set_title('Histogram Input') # menambah judul
```
